```python
def imprimir_valores(df_proce, columna):
  print(f"VALORES EN COLUMNA: {columna}")
  # Imprimir los valores únicos corregidos en la columna
  print(", ".join(f"{val}: {count}" for val, count in df_proce[columna].value_counts().items()))
```

```python
 # Creamos una función para convertir los strings de cada columna a numericos y remplazarlos por la moda
# Dado que son variables categoricas

import re

def procesar_columna(df_proce, columna):
    # Función para corregir valores de captura incorrectos en la columna
    def corregir_valor(valor):
        # Patrón de expresión regular para buscar valores incorrectos
        patron = r'\b(\d+)\D?\b'  # Este patrón busca cualquier secuencia de dígitos
        if isinstance(valor, str):
            # Eliminamos cualquier carácter que no sea un dígito
            valor_corregido = ''.join(filter(str.isdigit, valor))
            # Si el valor corregido no está vacío, devolvemos el entero
            if valor_corregido:
                return int(valor_corregido)
        return valor

    # Aplicar la corrección a la columna
    df_proce[columna] = df_proce[columna].apply(corregir_valor)

    df_proce[columna] = pd.to_numeric(df_proce[columna], errors='coerce')

    #obtener el recuento de valores únicos en la columna
    counts = df_proce[columna].value_counts()
    # Obtener la moda o el segundo valor más común, dependiendo de la columna
    if columna == "Proband_Relation":
        # Obtener los de valores únicos, excluyendo la moda
        no_moda = counts.drop(counts.idxmax())
        # Si aún hay valores después de excluir la moda,
        # tomar el más común de los restantes
        if not no_moda.empty:
            moda_columna = no_moda.idxmax()
        else:
            # Si todos los valores son iguales, usar la moda
            moda_columna = counts.idxmax()
    else:
        moda_columna = df_proce[columna].mode()[0]

    # Una vez que limpiamos de valores validos, sustituimos por la moda
    df_proce[columna] = df_proce[columna].fillna(moda_columna)

    # Convertir todos los valores a enteros
    df_proce[columna] = df_proce[columna].astype(int)

    return df_proce[columna]
```

```python
# Lista de nombres de las columnas categoricas que se van a sustituir los valores string por numericos
col_cat = ['Proband_Relation', 'Age', 'Marital_Status', 'School_Years', 'Employment']
print("\t -------- VALORES ANTES DE LA LIMPIEZA --------")
for columna in col_cat:
  imprimir_valores(df_copia, columna)
imprimir_valores(df_copia, 'Religion')
```

```python
def cuantificar_fuera_rango(df_proce, columna, rango_min, rango_max):
  # Contador para los valores fuera de rango
    valores_fuera_rango = 0
    # Función para obtener valores fuera de rango
    def cuantificar_valor(valor):
        nonlocal valores_fuera_rango
        # Verificar si el valor está fuera del rango permitido
        if valor < rango_min or valor > rango_max:
            valores_fuera_rango += 1
    df_proce[columna].apply(cuantificar_valor)
     # Imprimir la cantidad de valores fuera de rango
    print(f"Valores fuera de rango en columna {columna}: {valores_fuera_rango}")
```

```python
# Creamos una funcion para sustituir las variables categoricas que esten fuera de rango por la moda

def procesar_columna_range(df_pro, columna, rango_min, rango_max):
    # Corregir el valor para que esté dentro del rango
    def corregir_valor(valor):
        # Verificar si el valor está fuera del rango permitido
        if valor < rango_min or valor > rango_max:
            # Corregir el valor para que esté dentro del rango
            if valor < rango_min:
                return rango_min
            else:
                return rango_max
        else:
            return valor

    # Aplicar la corrección a la columna
    df_pro[columna] = df_pro[columna].apply(corregir_valor)

    return df_pro[columna]
```

```python
 # Definir los rangos permitidos para cada columna
rangos = {
    'Proband_Relation': (1,15),
    'Gender': (1,2),
    'Marital_Status': (1,6),
    'School_Years': (1,6),
    'Employment': (1,2),
    'Religion':(1,3)
}
for columna, (rango_min, rango_max) in rangos.items():
    cuantificar_fuera_rango(df_copia, columna, rango_min, rango_max)
```

```python
# Formateamos las columnas para que no esten fuera de rango
# Aplicamos la función procesar_columna_range a cada columna con su respectivo rango
for columna, (rango_min, rango_max) in rangos.items():
    df_copia[columna] = procesar_columna_range(df_copia, columna, rango_min, rango_max)
    cuantificar_fuera_rango(df_copia, columna, rango_min, rango_max)
imprimir_valores(df_copia, 'Religion')
```

Antes de comenzar, definiremos un orden específico para las relaciones en la base de datos, basándonos en la cercanía genética del probando con cada uno de sus familiares.

Así pues, definimos el conjunto de Familiares de primer grado como:

1.   Madre
2.   Padre
3.   Hermano
4.   Hermana
5.   Abuelos
6.   Hijos

Y de Segundo grado o lejanos:

1.   Tíos

```python
# Observamos las relaciones existentes en el conjunto de datos
print(multi_df['Proband_Relation'].unique())
```

```python
# Definimos el orden deseado de las relaciones familiares
relation_order = [
    "Proband","Mother", "Father", "Brother", "Sister", "Grandmother",
    "Grandfather", "Daughter", "Son", "Aunt", "Uncle", "Nephew", "Niece",
    "Cousin_Female", "Cousin_Male"
]

# Crear un diccionario para mapear relaciones a su respectivo orden
relation_mapping = {relation: i + 1 for i, relation in enumerate(relation_order)}

# Hacemos una copia de los datos
ordered_df = multi_df.copy()
# Mapeamos el orden de las relaciones por la columna de probandos
ordered_df['Relation_Order'] = ordered_df['Proband_Relation'].map(lambda x: relation_mapping.get(x, 0))
# Ordenar el DataFrame por la columna 'Order'
ordered_df = ordered_df.sort_values('Relation_Order').reset_index(drop=True)
ordered_df.drop(columns=['Proband_Relation'], inplace=True)
ordered_df.head()
```

Para trabajar con los métodos mencionados es necesario convertir nuestras variables categóricas en representaciones numéricas antes de calcular las correlaciones.

Usaremos 2 técnicas comunes para analizar nuestros datos y determinar cuál es el método más adecuado para nuestro caso de estudio:



1.   One Hot Encoding
2.   Label Encoding

**One Hot Encoding:**

El One-Hot Encoding es utilizado para manejar variables categóricas o de tipo nominal en un formato numérico, convierte variables categóricas en un formato numérico donde cada categoría se representa como una columna binaria (0 o 1), facilitando su uso en algoritmos de aprendizaje automático.


**Label Encoding: **

El Label Encoding es otra técnica para convertir variables categóricas en un formato numérico. Esta técnica es útil cuando existe un orden intrínseco en las categorías, como en el caso de etiquetas ordinales (por ejemplo, "bajo", "medio", "alto"), ya que conserva esta información de orden.

A cada categoría única se asigna a un número entero único. Por ejemplo, si tenemos las categorías "Rojo", "Verde" y "Azul", podríamos asignarles los números 0, 1 y 2 respectivamente.


```python
# Definimos las columnas categoricas
columnas_categoricas = ordered_df.select_dtypes(exclude=np.number).columns.tolist()
print("----- Columnas Categoricas -----")
columnas_categoricas
```

 Creamos una columna con valores binarios (0 o 1) para cada categoría. Esto es útil cuando no hay un orden implícito en las categorías. Por lo que utilizaremos las variables nominales


```python
from sklearn.preprocessing import OneHotEncoder

# Crear one-hot encoding para las variables categóricas
one_hot_df = ordered_df.copy()
# Remove Psychiatric_Diagnosis_DIGS
one_hot_df.drop(columns=['Psychiatric_Diagnosis_DIGS'], inplace=True)
# Quitar las columnas nominales de one_hot_df
one_hot_df.drop(columns=columnas_nominales, inplace=True)

# Inicializar el codificador OneHotEncoder
one_hot_encoder = OneHotEncoder(sparse=False)
# Ajustar y transformar las columnas nominales usando ordered_df
nominales_encoded = one_hot_encoder.fit_transform(ordered_df[columnas_nominales])
# Obtener los nombres de las nuevas columnas generadas por el codificador
nuevas_columnas = one_hot_encoder.get_feature_names_out(columnas_nominales)

# Convertir los valores codificados a enteros
nominales_encoded_int = nominales_encoded.astype(int)

print(nominales_encoded_int.shape, len(nuevas_columnas))
# Crear un nuevo DataFrame con las columnas codificadas
one_hot_encoded_df = pd.DataFrame(nominales_encoded_int, columns=nuevas_columnas)
one_hot_encoded_df = one_hot_encoded_df.astype(int)

# Resetear los índices para asegurar una alineación adecuada antes de hacer merge
one_hot_df.reset_index(drop=True, inplace=True)
one_hot_encoded_df.reset_index(drop=True, inplace=True)

# Concatenar el DataFrame codificado con las columnas del DataFrame original que no son nominales

one_hot_df = pd.merge(one_hot_df, one_hot_encoded_df, left_index=True, right_index=True, how='inner')
one_hot_df.head()
```

# Normalización
En nuestro análisis del conjunto de datos, creemos que no es necesario aplicar técnicas de normalización a las variables.

Esto se debe a que la mayoría de nuestras características son de naturaleza categórica, lo que significa que representan categorías discretas en lugar de valores numéricos continuos. Las variables categóricas, como el género, la religión, el estado civil y otras, no requieren normalización, ya que la comparación de sus valores se basa en la igualdad o diferencia entre categorías, en lugar de en la magnitud de los valores numéricos.

La única excepción es la variable de edad, que es numérica y podría necesitar ser escalada o normalizada si fuera relevante para nuestro análisis. Sin embargo en esta ocasión, no lo creemos necesario.

# ANALISIS PCA

```python
from sklearn.decomposition import PCA

def get_PCA(df_encoded):
  # Crear una instancia de PCA
  pca = PCA()

  # Ajustar el modelo PCA a nuestros datos
  pca.fit(df_encoded)

  # Visualizamos el número de componentes necesarios para explicar
  # cierta cantidad de varianza
  total_varianza_explicada = pca.explained_variance_ratio_.cumsum()
  varianza_noventa = sum(total_varianza_explicada < 0.90) + 1
  print("Número de componentes para explicar el 90% de la varianza:",
        varianza_noventa)

  pcsSummary = pd.DataFrame({'Standard deviation': np.sqrt(pca.explained_variance_),
                          'Proportion of variance': pca.explained_variance_ratio_,
                          'Cumulative proportion': np.cumsum(pca.explained_variance_ratio_)
                          })
  pcsSummary = pcsSummary.transpose()
  pcsSummary = pcsSummary.round(4)

  PC_components = np.arange(varianza_noventa) + 1

  _ = sns.set(style = 'whitegrid',
              font_scale = 1.2
              )

  fig, ax = plt.subplots(figsize=(10, 7))

  _ = sns.barplot(x = PC_components,
                  y = pca.explained_variance_ratio_[:varianza_noventa],
                  color = 'b'
                  )

  _ = sns.lineplot(x = PC_components-1,
                   y = np.cumsum(pca.explained_variance_ratio_[:varianza_noventa]),
                   color = 'black',
                   linestyle = '-',
                   linewidth = 2,
                   marker = 'o',
                   markersize = 8
                   )

  plt.title('Scree Plot > 90% Varianza')
  plt.xlabel('N-th Componente Principal')
  plt.ylabel('Varianza Explicada')
  plt.ylim(0, 1)
  plt.show()

  pca_Components_df = pd.DataFrame(pca.components_.transpose(),
                                columns=pcsSummary.columns,
                                index=df_encoded.columns
                                )
  pca_Components_df.columns = \
   ['PC' + str(i) for i in range(1, pca_Components_df.shape[1] + 1)]
  pca_Components_df = pca_Components_df.sort_values(by='PC1', ascending=False)
  return pca_Components_df.iloc[:, :varianza_noventa]
```
