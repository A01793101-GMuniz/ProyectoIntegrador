```python
def imprimir_valores(df_proce, columna):
  print(f"VALORES EN COLUMNA: {columna}")
  # Imprimir los valores únicos corregidos en la columna
  print(", ".join(f"{val}: {count}" for val, count in df_proce[columna].value_counts().items()))
```

```python
 # Creamos una función para convertir los strings de cada columna a numericos y remplazarlos por la moda
# Dado que son variables categoricas

import re

def procesar_columna(df_proce, columna):
    # Función para corregir valores de captura incorrectos en la columna
    def corregir_valor(valor):
        # Patrón de expresión regular para buscar valores incorrectos
        patron = r'\b(\d+)\D?\b'  # Este patrón busca cualquier secuencia de dígitos
        if isinstance(valor, str):
            # Eliminamos cualquier carácter que no sea un dígito
            valor_corregido = ''.join(filter(str.isdigit, valor))
            # Si el valor corregido no está vacío, devolvemos el entero
            if valor_corregido:
                return int(valor_corregido)
        return valor

    # Aplicar la corrección a la columna
    df_proce[columna] = df_proce[columna].apply(corregir_valor)

    df_proce[columna] = pd.to_numeric(df_proce[columna], errors='coerce')

    #obtener el recuento de valores únicos en la columna
    counts = df_proce[columna].value_counts()
    # Obtener la moda o el segundo valor más común, dependiendo de la columna
    if columna == "Proband_Relation":
        # Obtener los de valores únicos, excluyendo la moda
        no_moda = counts.drop(counts.idxmax())
        # Si aún hay valores después de excluir la moda,
        # tomar el más común de los restantes
        if not no_moda.empty:
            moda_columna = no_moda.idxmax()
        else:
            # Si todos los valores son iguales, usar la moda
            moda_columna = counts.idxmax()
    else:
        moda_columna = df_proce[columna].mode()[0]

    # Una vez que limpiamos de valores validos, sustituimos por la moda
    df_proce[columna] = df_proce[columna].fillna(moda_columna)

    # Convertir todos los valores a enteros
    df_proce[columna] = df_proce[columna].astype(int)

    return df_proce[columna]
```

```python
# Lista de nombres de las columnas categoricas que se van a sustituir los valores string por numericos
col_cat = ['Proband_Relation', 'Age', 'Marital_Status', 'School_Years', 'Employment']
print("\t -------- VALORES ANTES DE LA LIMPIEZA --------")
for columna in col_cat:
  imprimir_valores(df_copia, columna)
imprimir_valores(df_copia, 'Religion')
```

```python
def cuantificar_fuera_rango(df_proce, columna, rango_min, rango_max):
  # Contador para los valores fuera de rango
    valores_fuera_rango = 0
    # Función para obtener valores fuera de rango
    def cuantificar_valor(valor):
        nonlocal valores_fuera_rango
        # Verificar si el valor está fuera del rango permitido
        if valor < rango_min or valor > rango_max:
            valores_fuera_rango += 1
    df_proce[columna].apply(cuantificar_valor)
     # Imprimir la cantidad de valores fuera de rango
    print(f"Valores fuera de rango en columna {columna}: {valores_fuera_rango}")
```

```python
# Creamos una funcion para sustituir las variables categoricas que esten fuera de rango por la moda

def procesar_columna_range(df_pro, columna, rango_min, rango_max):
    # Corregir el valor para que esté dentro del rango
    def corregir_valor(valor):
        # Verificar si el valor está fuera del rango permitido
        if valor < rango_min or valor > rango_max:
            # Corregir el valor para que esté dentro del rango
            if valor < rango_min:
                return rango_min
            else:
                return rango_max
        else:
            return valor

    # Aplicar la corrección a la columna
    df_pro[columna] = df_pro[columna].apply(corregir_valor)

    return df_pro[columna]
```

```python
 # Definir los rangos permitidos para cada columna
rangos = {
    'Proband_Relation': (1,15),
    'Gender': (1,2),
    'Marital_Status': (1,6),
    'School_Years': (1,6),
    'Employment': (1,2),
    'Religion':(1,3)
}
for columna, (rango_min, rango_max) in rangos.items():
    cuantificar_fuera_rango(df_copia, columna, rango_min, rango_max)
```

```python
# Formateamos las columnas para que no esten fuera de rango
# Aplicamos la función procesar_columna_range a cada columna con su respectivo rango
for columna, (rango_min, rango_max) in rangos.items():
    df_copia[columna] = procesar_columna_range(df_copia, columna, rango_min, rango_max)
    cuantificar_fuera_rango(df_copia, columna, rango_min, rango_max)
imprimir_valores(df_copia, 'Religion')
```

### Mapear Relaciones Familiares

Antes de agrupar, definiremos un orden específico para las relaciones en la base de datos, basándonos en la cercanía genética del probando con cada uno de sus familiares.

Así pues, definimos el conjunto de Familiares de primer grado como:

1.   Madre
2.   Padre
3.   Hermano
4.   Hermana
5.   Abuelos
6.   Hijos

Y de Segundo grado o lejanos:

1.   Tíos
2.   Sobrinos
3.   Primos

```python
# Observamos las relaciones existentes en el conjunto de datos
print(df_copia['Proband_Relation'].unique())
```

```python
# Definimos el orden deseado de las relaciones familiares
relation_order = [
    "Proband","Mother", "Father", "Brother", "Sister", "Grandmother",
    "Grandfather", "Daughter", "Son", "Aunt", "Uncle", "Nephew", "Niece",
    "Cousin_Female", "Cousin_Male"
]

# Crear un diccionario para mapear relaciones a su respectivo orden
relation_mapping = {relation: i + 1 for i, relation in enumerate(relation_order)}

# Hacemos una copia de los datos
ordered_df = df_copia.copy()
# Mapeamos el orden de las relaciones por la columna de probandos
ordered_df['Relation_Order'] = ordered_df['Proband_Relation'].map(lambda x: relation_mapping.get(x, 0))
# Ordenar el DataFrame por la columna 'Order'
ordered_df = ordered_df.sort_values('Relation_Order').reset_index(drop=True)
ordered_df.head()
```

```python
# Agrupar por Family_ID y combinar los valores de cada columna en una lista
grouped_df = df_sinproband.groupby('Family_ID').agg(lambda x: x.tolist()).reset_index()
grouped_df.head()
```

```python
grouped_df.shape
```

```python
# Unimos los datos de los Familiares con los del probando de acuerdo a su Family ID
merged_df = pd.merge(df_proband.drop(columns=['index', 'Proband_Relation']), grouped_df, on='Family_ID', how='left')
```

```python
# Lista de columnas de cada familiar
familiar_columns = list(merged_df.columns)
familiar_columns.remove("Family_ID")
# Identificar las columnas que contienen "_x" en su nombre
proband_columns = [col for col in familiar_columns if '_x' in col]
# Eliminar las columnas de proband_columns de familiar_columns
familiar_columns = [col for col in familiar_columns if col not in proband_columns]

# Creamos una lista para almacenar los valores de cada familia
data_rows = []

# Iteramos sobre cada fila del DataFrame
for index, row in merged_df.iterrows():
    family_id = row['Family_ID']
    if isinstance(row['Age_y'], list):
      num_families = min(2, len(row['Age_y']))  # Tomamos solo los datos de los primeros 2 familiares
      if num_families >= 2:
        new_row = {'Family_ID': family_id}
        # Agregar los datos de las columnas de proband_columns a new_row
        for col in proband_columns:
          new_row[col.replace("_x", "_Proband")] = row[col]

        for i in range(num_families):
          for col in familiar_columns:
            new_row[f"{col}_Familiar{i+1}"] = row[col][i]
        data_rows.append(new_row)

# Creamos un nuevo DataFrame con los datos reorganizados
new_df = pd.DataFrame(data_rows)
```

### Rescatar Familiares Adicionales

```python
# Crear los dos arreglos utilizando slice
orden1 = relation_order[1:9]
orden2 = relation_order[9:]

# Mostramos los grupos familiares
print("Orden 1:", orden1)
print("Orden 2:", orden2)
```

```python
addFamColumnas = [
    "Psychosis_DIGS",
    "D+AH1epression_DIGS",
    "Anxiety_DIGS",
    "Eating_Disorders_DIGS",
    "Alcohol_Abuse/Dependence_DIGS",
    "Drugs_Abuse/Dependence_DIGS",
    "Behaviour_Disorders",
    "Personality_Disorders"
]
```

```python
# Creamos un dataframe para el XOR de familiares adicionales
new_columns_cercanos = [col + "_Cercanos" for col in addFamColumnas]
new_columns_lejanos = [col + "_Lejanos" for col in addFamColumnas]

famAdicional_df = new_df[['Family_ID']].copy()
# Crear DataFrames de ceros para las nuevas columnas
df_cercanos = pd.DataFrame(False, index=new_df.index, columns=new_columns_cercanos)
df_lejanos = pd.DataFrame(False, index=new_df.index, columns=new_columns_lejanos)

# Concatenar family_ID, cercanos y lejanos en el nuevo DataFrame
famAdicional_df = pd.concat([famAdicional_df, df_cercanos, df_lejanos], axis=1)

famAdicional_df.head(10)
```

```python
# Función para eliminar los primeros dos elementos de una lista,
# Los cuales representaran los 2  familiares mas cercanos al probando
def remove_first_two_elements(lst):
    if len(lst) > 2:
        return lst[2:]
    else:
        return []

restOfFams = grouped_df.copy()
# Eliminar los primeros dos elementos de cada lista en todas las columnas
restOfFams.iloc[:, 1:] = restOfFams.iloc[:, 1:].applymap(remove_first_two_elements)

# Filtrar las filas que tengan al menos un elemento mas en cada columna
restOfFams = restOfFams[restOfFams.iloc[:, 1:].applymap(lambda x: len(x) >= 1).all(axis=1)]

# Mostrarmos los primeros registros del dataframe con Familiares adicionales
restOfFams.head()
```

```python
# Mostramos la composicion de registros y features en nuestro nuevo DF
# de familiares restantes
restOfFams.shape
```

```python
# Invertimos el mapeo de relaciones para poder usarlo posteriormente con el DF
relation_mapping =  {v: k for k, v in relation_mapping.items()}
relation_mapping
```

```python
def separar_familiares(row):
  # Creamos una lista para almacenar los valores de cada familia
  count_O1 = 0
  count_O2 = 0
  for ord in row['Relation_Order']:
    if relation_mapping[ord] in orden1:
      # Cuenta de familiares cercanos
      count_O1 += 1
    elif relation_mapping[ord] in orden2:
      # Cuenta de familiares lejanos
      count_O2 += 1
  if count_O1 == 0 and count_O2 == 0:
    return
  for col in addFamColumnas:
    # Obtener el índice de la fila en famAdicional_df que tiene el mismo Family_ID que row
    indice_fila = famAdicional_df.index[famAdicional_df["Family_ID"] == row["Family_ID"]]
    if count_O1 > 0 and row[col][0:count_O1].count("Yes") > 0:
      if not indice_fila.empty:
        for ind in indice_fila:
          # Modificar el valor en la columna 'Nombre_Columna' en la fila correspondiente
          famAdicional_df.at[ind, f"{col}_Cercanos"] = True
    if count_O2 > 0 and row[col][count_O1:].count("Yes") > 0:
      if not indice_fila.empty:
        for ind in indice_fila:
          # Modificar el valor en la columna 'Nombre_Columna' en la fila correspondiente
          famAdicional_df.at[ind, f"{col}_Lejanos"] = True

# Aplicar la función a cada fila del DataFrame
restOfFams.apply(separar_familiares, axis=1)
```
